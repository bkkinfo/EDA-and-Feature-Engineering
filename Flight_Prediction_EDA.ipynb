## Feature Engineering process

spliting Date_of_Journey to DMY format

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]  #.apply(lambda x:x.split("/")[0])
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1] #.apply(lambda x:x.split("/")[1])
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]  #.apply(lambda x:x.split("/")[2])
```

```python
final_df.head()
```

```python
final_df=final_df.drop('Date_of_Journey', axis=1)
```

```python
final_df.info()
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hours'] = final_df['Arrival_Time'].apply(lambda x : x.split(":")[0])
final_df['Arrival_min'] = final_df['Arrival_Time'].apply(lambda x : x.split(":")[1])
```

```python
final_df['Arrival_hours']=final_df['Arrival_hours'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df['Dep_hours'] = final_df['Dep_Time'].apply(lambda x : x.split(":")[0])
final_df['Dep_mins'] = final_df['Dep_Time'].apply(lambda x : x.split(":")[1])
```

```python
final_df['Dep_hours']=final_df['Dep_hours'].astype(int)
final_df['Dep_mins']=final_df['Dep_mins'].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```
